```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data
```

## Player Count

* Display the total number of players


```python
total_players=len(purchase_data.SN.unique())
#print(f"Total number of players: {total_players}")
pd.DataFrame({
    "Total players:":total_players
}, index=[0])
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
number_of_unique_items=len(purchase_data["Item ID"].unique())
average_price=purchase_data["Price"].mean()
number_of_purchases=purchase_data["Purchase ID"].count()
total_revenue=purchase_data["Price"].sum()
dict={"Number of unique items":number_of_unique_items,
      "Average Price":average_price,
     "Number of Purchases": number_of_purchases,
     "Tota Revenue":total_revenue}
summary_table=pd.DataFrame(dict,index=[0])
summary_table=summary_table.astype(float)
summary_table["Number of unique items"]=summary_table["Number of unique items"].map("{:,.0f}".format)
summary_table["Average Price"]=summary_table["Average Price"].map("${:,.2f}".format)
summary_table["Number of Purchases"]=summary_table["Number of Purchases"].map("{:,.0f}".format)
summary_table["Tota Revenue"]=summary_table["Tota Revenue"].map("${:,.2f}".format)
summary_table
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
total=purchase_data.SN.nunique()
total_count=purchase_data.groupby("Gender").SN.nunique()
percentage=purchase_data.groupby("Gender").SN.nunique()/total*100
summary_table_gender_demographics=pd.DataFrame({
    "Purchase Count": total_count, "Percentage":percentage
})  
summary_table_gender_demographics.Percentage=summary_table_gender_demographics.Percentage.map("%{:.2f}".format)
summary_table_gender_demographics
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
purchase_data=purchase_data.rename(columns={"Purchase ID": "purchase_id"})
purchase_count=purchase_data.groupby("Gender").purchase_id.nunique()
avg_purchase_price=purchase_data.groupby("Gender").Price.mean()
total_purchase_value=purchase_data.groupby("Gender").Price.sum()
avg_total_purchase_person=purchase_data.groupby("Gender").Price.sum()/total_count
summary_table_purchase_analysis=pd.DataFrame({
    "Purchase Count": purchase_count, 
    "Average Purchase Price":avg_purchase_price,
    "Total Purchase Value": total_purchase_value,
    "Avg Total Purchase per Person": avg_total_purchase_person
})  

summary_table_purchase_analysis["Purchase Count"]=summary_table_purchase_analysis["Purchase Count"].map("{:.0f}".format)
summary_table_purchase_analysis["Average Purchase Price"]=summary_table_purchase_analysis["Average Purchase Price"].map("${:.2f}".format)
summary_table_purchase_analysis["Total Purchase Value"]=summary_table_purchase_analysis["Total Purchase Value"].map("${:.2f}".format)
summary_table_purchase_analysis["Avg Total Purchase per Person"]=summary_table_purchase_analysis["Avg Total Purchase per Person"].map("${:.2f}".format)
summary_table_purchase_analysis
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
group_names=["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]
bins=[0,9,14,19,24,29,34,39,110]
purchase_data["age_group"]=pd.cut(purchase_data.Age, bins,labels=group_names, include_lowest=True)
total_count_age=purchase_data.groupby("age_group").SN.nunique()
percentage_age=purchase_data.groupby("age_group").SN.nunique()/total*100
summary_table_age_demographics=pd.DataFrame({
    "Total Count": total_count_age, "Percentage of Players":percentage_age
}) 
summary_table_age_demographics["Percentage of Players"]=summary_table_age_demographics["Percentage of Players"].map("%{:.2f}".format)
summary_table_age_demographics
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
total_count_age_group=purchase_data.groupby("age_group").SN.nunique()
purchase_count_age=purchase_data.groupby("age_group").purchase_id.nunique()
avg_purchase_price_age=purchase_data.groupby("age_group").Price.mean()
total_purchase_value_age=purchase_data.groupby("age_group").Price.sum()
avg_total_purchase_person_age=purchase_data.groupby("age_group").Price.sum()/total_count_age_group
summary_table_purchase_analysis_age=pd.DataFrame({
    "Purchase Count": purchase_count_age, 
    "Average Purchase Price":avg_purchase_price_age,
    "Total Purchase Value": total_purchase_value_age,
    "Avg Total Purchase per Person": avg_total_purchase_person_age
})  
summary_table_purchase_analysis_age=summary_table_purchase_analysis_age.astype(float)
summary_table_purchase_analysis_age["Purchase Count"]=summary_table_purchase_analysis_age["Purchase Count"].map("{:.0f}".format)
summary_table_purchase_analysis_age["Average Purchase Price"]=summary_table_purchase_analysis_age["Average Purchase Price"].map("${:.2f}".format)
summary_table_purchase_analysis_age["Total Purchase Value"]=summary_table_purchase_analysis_age["Total Purchase Value"].map("${:.2f}".format)
summary_table_purchase_analysis_age["Avg Total Purchase per Person"]=summary_table_purchase_analysis_age["Avg Total Purchase per Person"].map("${:.2f}".format)
summary_table_purchase_analysis_age
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
total_count_sn=purchase_data.groupby("SN").SN.nunique()
purchase_count_sn=purchase_data.groupby("SN").purchase_id.nunique()
avg_purchase_price_sn=purchase_data.groupby("SN").Price.mean()
total_purchase_value_sn=purchase_data.groupby("SN").Price.sum()

summary_table_purchase_analysis_sn=pd.DataFrame({
    "Purchase Count": purchase_count_sn, 
    "Average Purchase Price":avg_purchase_price_sn,
    "Total Purchase Value": total_purchase_value_sn
})  
summary_table_purchase_analysis_sn=summary_table_purchase_analysis_sn.sort_values("Total Purchase Value",ascending=False)
summary_table_purchase_analysis_sn=summary_table_purchase_analysis_sn.head()

summary_table_purchase_analysis_sn["Purchase Count"]=summary_table_purchase_analysis_sn["Purchase Count"].map("{:.0f}".format)
summary_table_purchase_analysis_sn["Average Purchase Price"]=summary_table_purchase_analysis_sn["Average Purchase Price"].map("${:.2f}".format)
summary_table_purchase_analysis_sn["Total Purchase Value"]=summary_table_purchase_analysis_sn["Total Purchase Value"].map("${:.2f}".format)


summary_table_purchase_analysis_sn
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results
